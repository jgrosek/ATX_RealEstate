```python
from homeharvest import scrape_property
import pandas as pd

pd.set_option('display.max_columns', None)  # Show all columns
pd.set_option('display.max_rows', None)     # Show all rows
pd.set_option('display.width', None)        # Auto-adjust display width to fit console
pd.set_option('display.max_colwidth', 50)   # Limit max column width to 50 characters
```

# scrapes all 3 sites by default
scrape_property(
    location="Austin",
    listing_type="for_rent"
)

```python
# check rentals
rent = scrape_property(
    location="Austin",
    site_name=["redfin","zillow"],
    listing_type="for_rent"
)
rent.to_csv('austin-rentals.csv',header=True)

# check sales
buy = scrape_property(
    location="Austin",
    site_name=["redfin","zillow"],
    listing_type="for_sale"
)
buy.to_csv('austin-purchases.csv',header=True)
```

```python
buy.head()
```

```python
# Group the data by "property_type" and count the occurrences of each type
property_type_counts = buy.groupby('property_type').size().reset_index(name='count')

# Sort the DataFrame in descending order by the "count" column
property_type_counts = property_type_counts.sort_values(by='count', ascending=False)

# Calculate the total count of all property types
total_count = property_type_counts['count'].sum()

# Add a new column with the percentage of each property type
property_type_counts['percentage'] = (property_type_counts['count'] / total_count) * 100

# Round the "percentage" column to 1 decimal point
property_type_counts['percentage'] = property_type_counts['percentage'].round(1)

print(property_type_counts)
```
